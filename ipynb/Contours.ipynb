```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"C:/Users/kandu/Desktop/All sub/ComputerVision/images/contourImg.jpg")
cv2.imshow("Original Image", img)
cv2.waitKey()

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Image", gray_img)
cv2.waitKey()

canny_edges = cv2.Canny(gray_img, 100, 120)
cv2.imshow("Canny Edges", canny_edges)
cv2.waitKey()

contours, hierarchy = cv2.findContours(canny_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print("Number of contours: ", len(contours))
print(f"Contours: {contours}")

cv2.drawContours(img, contours, -1, (166,0,123), 4) # cv2.drawContours(image, contours, contourIdx, color, thickness) -> -1 indicates draw all contours at once
cv2.imshow("drawing contours ", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
